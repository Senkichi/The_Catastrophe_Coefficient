```python
from sklearn import tree
import pandas as pd
import os
```

```python
data = pd.read_csv('../data/2050_rcp85_bc_frame.csv') 
```

```python
# data.drop(['Unnamed: 0'], axis = 1, inplace = True)
data.dropna(axis = 0, how = 'any', inplace = True)
data.rename(columns={'bc85bi501': 'Annual Mean Temperature', 
                     'bc85bi5010': 'Mean Diurnal Range (Mean of monthly (max temp - min temp))',
                     'bc85bi5011': 'Isothermality',
                     'bc85bi5012': 'Temperature Seasonality',
                     'bc85bi5013': 'Max Temperature of Warmest Month',
                     'bc85bi5014': 'Min Temperature of Coldest Month',
                     'bc85bi5015': 'Temperature Annual Range',
                     'bc85bi5016': 'Mean Temperature of Wettest Quarter',
                     'bc85bi5017': 'Mean Temperature of Driest Quarter',
                     'bc85bi5018': 'Mean Temperature of Warmest Quarter',
                     'bc85bi5019': 'Mean Temperature of Coldest Quarter',
                     'bc85bi502': 'Annual Precipitation',
                     'bc85bi503': 'Precipitation of Wettest Month',
                     'bc85bi504': 'Precipitation of Driest Month',
                     'bc85bi505': 'Precipitation Seasonality (Coefficient of Variation)',
                     'bc85bi506': 'Precipitation of Wettest Quarter',
                     'bc85bi507': 'Precipitation of Driest Quarter',
                     'bc85bi508': 'Precipitation of Warmest Quarter',
                     'bc85bi509': 'Precipitation of Coldest Quarter',
                    }, inplace=True)

# data.columns
data.drop(['Unnamed: 0', 
           'Max Temperature of Warmest Month', 
           'Min Temperature of Coldest Month', 
           'Precipitation of Wettest Month', 
           'Precipitation of Driest Month', 
           'Precipitation Seasonality (Coefficient of Variation)', 
           'Mean Diurnal Range (Mean of monthly (max temp - min temp))', 
           'Temperature Annual Range', 'Annual Mean Temperature', 
           'Isothermality', 
           'Temperature Seasonality', 
           'Annual Precipitation'], axis=1, inplace=True)
data.head()
```

```python
data.dtypes
data['Mean Temperature of Wettest Quarter'] = data['Mean Temperature of Wettest Quarter'].astype(float)
data['Mean Temperature of Driest Quarter'] = data['Mean Temperature of Driest Quarter'].astype(float)
data['Mean Temperature of Warmest Quarter'] = data['Mean Temperature of Warmest Quarter'].astype(float)
data['Mean Temperature of Coldest Quarter'] = data['Mean Temperature of Coldest Quarter'].astype(float)
data['Precipitation of Wettest Quarter'] = data['Precipitation of Wettest Quarter'].astype(float)
data['Precipitation of Driest Quarter'] = data['Precipitation of Driest Quarter'].astype(float)
data['Precipitation of Warmest Quarter'] = data['Precipitation of Warmest Quarter'].astype(float)
data['Precipitation of Coldest Quarter'] = data['Precipitation of Coldest Quarter'].astype(float)

data.dtypes
```

```python
new_data = data.drop(['x','y'], axis=1)
new_data.head()
```

```python

from sklearn.preprocessing import LabelEncoder, StandardScaler

scaler = StandardScaler().fit(new_data)

new_data_scaled = scaler.transform(new_data)
```

```python
from keras.models import load_model

# model.save('CNNSequential.h5')

model = load_model('CNNSequential.h5')
```

```python
prediction = model.predict(new_data_scaled)

print(f'Prediction: {prediction[:10]}')
```

```python
count=0
for p in prediction:
    if p[0]>0:
        count +=1
print(f'Prediction count > 0: {count}')
print(len(prediction))
print(len(new_data))
```
